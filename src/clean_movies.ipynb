```python
# Import the libraries
import pandas as pd
import string
```

```python
# Read the CSV file
path_in = r'/content/IMDB Top 250 Movies.csv'
movies = pd.read_csv(path_in, index_col=False)
```

```python
# Display the movie dataset 
movies
```

```python
# rename some of the header
movies = movies.rename(columns={'rank': 'Rank',
                            'name': 'Name',
                            'year': 'Year',
                            'rating': 'Rating',
                            'genre': 'Genres',
                            'certificate': 'Certificate',
                            'run_time': 'Run_time',
                            'tagline': 'Tagline',
                            'budget': 'Budget',
                            'box_office': 'Box_office',
                            'casts': 'Casts',
                            'directors': 'Directors',
                            'writers': 'Writers'})
```

```python
# Show list of column name of the dataset
movies.columns.tolist()
```

```python
# Get the information of the dataset
movies.info()
```

```python
# Calculate the occurences of "Not Available" value in each column
count_NA = movies.apply(lambda x: x.value_counts().get('Not Available', 0))
print(count_NA)
```

```python
# Change the 'Not Available' variable in the certificate column to 'Unknown' and clean other cells
def change_cell(cell):
  if (cell != 'Not Available'):
    value = cell
  else:
    value = 'Unknown'
  return value

def certificate_transform(row):
  row.Certificate = change_cell(row.Certificate)
  return row
    
movies = movies.apply(certificate_transform, axis=1)
```

```python
# Change the 'Not Available' variable in the run_time column to 'Unknown' and convert the length of the movie to minutes
def replace_unit (cell):
  if (cell != 'Not Available'):
    time = cell.split()
    if (len(time) > 1):
      total_minutes = int(time[0].strip('h')) * 60 + int(time[1].strip('m'))
    elif (time[0][-1] == 'h'):
      total_minutes = int(time[0].strip('h')) * 60
    elif (time[0][-1] == 'm'):
      total_minutes = int(time[0].strip('m'))
    else:
      total_minutes = 0
  else:
    total_minutes = 'Unknown'
  return total_minutes

def run_time_transform(row):
    row.Run_time = replace_unit(row.Run_time)
    return row
    
movies = movies.apply(run_time_transform, axis=1)
```

```python
# Change the 'Not Available' variable in the budget column to 'Unknown' and clean other cells
def cell_replace(unit):
  if (unit != 'Not Available'):
    values = unit.split()
    if (len(values) > 1):
      if (values[0] == 'EM'):
        value = values[1]
      elif (values[0] == 'RF'):
        value = values[1]
    else:
      value = values[0].strip('$')
  else:
    value = 'Unknown'
  return value

def budget_transform(row):
    row.Budget = cell_replace(row.Budget)
    return row
    
movies = movies.apply(budget_transform, axis=1)
```

```python
# Change the 'Not Available' variable in the box_office column to 'Unknown' and clean other cells
def cell_change(cell):
  if (cell != 'Not Available'):
    values = cell.split()
    value = values[0]
  else:
    value = 'Unknown'
  return value

def box_office_transform(row):
  row.Box_office = cell_change(row.Box_office)
  return row
    
movies = movies.apply(box_office_transform, axis=1)
```

```python
# Calculate the occurences of 'Not Available' value in each column
count_NA = movies.apply(lambda x: x.value_counts().get('Not Available', 0))
print(count_NA)
```

```python
# write the changed dataset to a new file
movies.to_csv('/content/clean_movies.csv', index=False)
```
